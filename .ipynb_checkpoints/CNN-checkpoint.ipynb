```python
import torch
from torch import nn, optim
from jcopdl.callback import Callback, set_config
```

```python
device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
device
```

# Dataset & Dataloader

```python
from torchvision import transforms, datasets
from torch.utils.data import DataLoader
```

```python
bs = 128
crop_size = 64 #seharusnya tidak sekecil ini biasanya 224

train_transform = transforms.Compose([
    transforms.RandomRotation(15), # 15derajat
    transforms.RandomResizedCrop(crop_size, scale=(0.8, 1.0)), #maksimal zoomin 80%
    transforms.RandomHorizontalFlip(), # gambarnya dibalik balik
    transforms.ToTensor() # konvert to tensor
])

# datatest jangan diaugment, cuma ttp harus ukuranya sesuai input
# kalo cropsizenya 64 kita akan lebih gede sedikit dalam ngerisize
# kenapa perlu dirisize?? jadi image kita diresize dlu jadi aga kecil
# aga mirip dngan yg akan kita crop, baru kita center crop 64 * 64
# jadi dapet informasinya, kalo gadrize nanti jatohnya kan kecil bgt dr gambar aslinya dan kita gadapet apa apa, gt intiinya

test_transform = transforms.Compose([
    transforms.Resize(70),
    transforms.CenterCrop(crop_size),
    transforms.ToTensor()
])



train_set = datasets.ImageFolder("data/train/", transform=train_transform)
trainloader = DataLoader(train_set, shuffle=True, batch_size=bs, num_workers=4)

test_set = datasets.ImageFolder("data/test/", transform=test_transform)
testloader = DataLoader(test_set, shuffle=True, batch_size=bs)
```

# Arsitektur & Config

```python
Image("data/cnn2.png")
```

```python
from jcopdl.layers import conv_block, linear_block
```

```python
class CNN(nn.Module):
    def __init__(self):
        super().__init__()
        self.conv = nn.Sequential(
            conv_block(3, 8),
            conv_block(8, 16),
            conv_block(16, 32),
            conv_block(32, 64),
            nn.Flatten()
        )
        
        self.fc = nn.Sequential(
            linear_block(1024, 256, dropout=0.1),
            linear_block(256, 2, activation="lsoftmax")
        )
        
    def forward(self, x):
        x = self.conv(x)
        x = self.fc(x)
        return x
         
```

```python
config = set_config({
    "batch_size" : bs,
    "crop_size" : crop_size
})
```

# Training preparation -> MCOC

```python
model = CNN().to(device)
criterion = nn.NLLLoss()
optimizer = optim.AdamW(model.parameters(), lr=0.001)
callback = Callback(model, config, outdir="model")
```

```python
from tqdm.auto import tqdm

def loop_fn(mode, dataset, dataloader, model, criterion, optimizer, device):
    if mode == "train":
        model.train()
    elif mode == "test":
        model.eval() 
    cost = correct = 0 
    for feature, target in tqdm(dataloader, desc=mode.title()): # custsome description
        feature, target = feature.to(device), target.to(device)
        output = model(feature) #feed forward
        loss = criterion(output, target)
        
        if mode == "train":
            loss.backward() 
            optimizer.step()
            optimizer.zero_grad()
            
        cost += loss.item() * feature.shape[0]
        correct += (output.argmax(1) == target).sum().item()
    cost = cost / len(dataset)
    acc = correct / len(dataset)
    return cost, acc
```

```python
   
while True:
    train_cost, train_score = loop_fn("train", train_set, trainloader, model, criterion, optimizer, device)
    with torch.no_grad():
         test_cost, test_score = loop_fn("test", test_set, testloader, model, criterion, optimizer, device)
         
    # Logging and Reporting
    callback.log(train_cost, test_cost, train_score, test_score)

    # Checkpoint
    callback.save_checkpoint()

    # Runtime Plotting
    callback.cost_runtime_plotting()
    callback.score_runtime_plotting()

    # Early Stopping
    if callback.early_stopping(model, monitor='test_score'):
        callback.plot_cost()
        callback.plot_score()
        break
    
```
